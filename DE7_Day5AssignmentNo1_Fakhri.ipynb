```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import time
import csv

# Set up WebDriver
chrome_options = Options()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')

# Membuka browser di Google Colab menggunakan ChromeDriver
driver = webdriver.Chrome(options=chrome_options)

# URL WebScraper test site (e-commerce products)
url = 'https://webscraper.io/test-sites/e-commerce/allinone'

# Akses halaman WebScraper.io
driver.get(url)

# Tunggu beberapa detik agar halaman sepenuhnya dimuat
time.sleep(5)

# Ambil konten halaman setelah dimuat
html = driver.page_source

# Parsing HTML menggunakan BeautifulSoup
soup = BeautifulSoup(html, 'html.parser')

# Mencari elemen yang berisi daftar produk
products = soup.find_all('div', class_='thumbnail')

# Loop untuk mengambil nama, deskripsi, dan harga produk
for product in products:
    # Mengambil nama produk
    product_name = product.find('a', class_='title').text.strip()

    # Mengambil deskripsi produk
    product_description = product.find('p', class_='description').text.strip()

    # Mengambil harga produk
    product_price = product.find('h4', class_='price').text.strip()

    # Menampilkan nama, deskripsi, dan harga produk
    print(f'Nama Produk: {product_name}, Deskripsi: {product_description}, Harga: {product_price}')

  # Buka atau buat file CSV untuk menyimpan hasil scraping
with open('produk_webscraper.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)

    # Menulis header ke file CSV
    writer.writerow(['Nama Produk', 'Deskripsi', 'Harga'])

    # Loop untuk mengambil nama, deskripsi, dan harga produk
    for product in products:
        # Mengambil nama produk
        product_name = product.find('a', class_='title').text.strip()

        # Mengambil deskripsi produk
        product_description = product.find('p', class_='description').text.strip()

        # Mengambil harga produk
        product_price = product.find('h4', class_='price').text.strip()

        # Menulis data ke file CSV
        writer.writerow([product_name, product_description, product_price])

# Menutup browser setelah selesai
driver.quit()

print("Data produk telah disimpan ke file produk_webscraper.csv")
```
